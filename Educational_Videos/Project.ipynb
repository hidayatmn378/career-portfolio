```python
from manim import *

def FirstScene(scene : Scene):
       # animate stuff
        text1=Text("Menghitung Integral Luas - Aplikasi Integral", font_size=40, weight=BOLD)
        text1.set_color_by_gradient(YELLOW, RED)
        text2=Text("Author: Muh. Nur Hidayat \n NPM: 2108201010005", font_size=40)
        text2[0:6].set_color(YELLOW)
        text2[0:6].set_color(YELLOW)
        text2[21:24].set_color(YELLOW)
        
        rectangle = SurroundingRectangle(text1, color=RED, buff=0.1)
        
        scene.play(Create(rectangle))
        scene.play(Write(text1),run_time=5.5)
        scene.play(Unwrite(text1, reverse=True))
        scene.play(Uncreate(rectangle))
        scene.play(Write(text2),run_time=5.5)
        scene.play(Unwrite(text2, reverse=False))
        scene.wait()

def SecondScene(scene : Scene):
      # animate stuff
        paragraphe = Paragraph("1. Luas Di bawah Kurva - Persegi\n",
                              "2. Luas Di bawah Kurva - Trapesium\n",
                              "3. Integral Luas terhadap Sumbu-X\n",
                              "4. Integral Riemann\n",
                              "5. Contoh Soal - Pembahasan")
        paragraphe.scale(0.6)
        paragraphe.to_corner(UL) 
        
        underline1 = Underline(paragraphe[0:1], color=RED, buff=0.1)
        underline2 = Underline(paragraphe[1:4], color=RED, buff=0.1)
        underline3 = Underline(paragraphe[4:6], color=RED, buff=0.1)
        underline4 = Underline(paragraphe[6:8], color=RED, buff=0.1)
        underline5 = Underline(paragraphe[8:15], color=RED, buff=0.1)
        
        scene.play(Write(paragraphe))
        scene.play(GrowFromCenter(underline1), run_time=1.5)
        scene.wait()
        scene.play(GrowFromCenter(underline2), run_time=1.5)
        scene.wait()
        scene.play(GrowFromCenter(underline3), run_time=1.5)
        scene.wait()
        scene.play(GrowFromCenter(underline4), run_time=1.5)
        scene.wait()
        scene.play(GrowFromCenter(underline5), run_time=1.5)
        scene.wait(3)
        
        scene.play(Uncreate(underline5))
        scene.play(Uncreate(underline4))
        scene.play(Uncreate(underline3))
        scene.play(Uncreate(underline2))
        scene.play(Uncreate(underline1))
        scene.play(Unwrite(paragraphe))
        
def ThirdScene(scene : Scene):
        ax = Axes(
            x_range=[0, 6],
            y_range=[0, 6],
            x_axis_config={"numbers_to_include": [2, 5]},
            tips=False,
            y_axis_config={"numbers_to_include": [0, 5]},
        )

        labels = ax.get_axis_labels()

        curve_1 = ax.plot(lambda x: 5, x_range=[0, 6], color=BLUE_C)
        curve_2 = ax.plot(
            lambda x: 0,
            x_range=[0, 6],
            color=GREEN_B,
        )
        
        t_label = ax.get_graph_label(
            graph=curve_1,
            label= MathTex(r"y=5"),
            x_val=6,
            dot=False,
            direction=UR)
        
        line_1 = ax.get_vertical_line(ax.input_to_graph_point(2, curve_1), color=YELLOW)
        line_2 = ax.get_vertical_line(ax.i2gp(5, curve_1), color=YELLOW)

#         riemann_area = ax.get_riemann_rectangles(curve_1, x_range=[0.3, 0.6], dx=0.03, color=BLUE, fill_opacity=0.5)
        area = ax.get_area(curve_2, [2, 5], bounded_graph=curve_1, color=GREY, opacity=0.5)
    
        transform_title = Tex("Hitung luas daerah yang diarsir?", font_size=72)
        transform_title.to_corner(UP + RIGHT)
        
        hasil_title =MathTex(
            "L = p \\times l = 3 \\times 5 = 15 \\text{ satuan luas}"
        )
        scene.add(ax, labels, transform_title)
        
        scene.play(Create(curve_1),run_time=2)
        scene.wait()
        scene.add(t_label)
        scene.play(Create(curve_2),run_time=2)
        scene.wait()
        scene.play(Create(line_1),run_time=0.5)
        scene.wait()
        scene.play(Create(line_2),run_time=0.5)
        scene.wait()
        scene.play(Create(area),run_time=2.5)
        scene.wait(3)
        scene.play(Create(hasil_title),run_time=5.5)
        scene.wait(3)
        
        scene.play(Uncreate(hasil_title))
        scene.play(Uncreate(area))
        scene.play(Uncreate(line_1))
        scene.play(Uncreate(line_2))
        scene.play(Uncreate(curve_1))
        scene.play(Uncreate(curve_2))
        scene.remove(ax, labels, t_label, transform_title)             
        scene.wait(2)
        
def FourthScene(scene : Scene):
        ax = Axes(
            x_range=[0, 6],
            y_range=[0, 12],
            x_axis_config={"numbers_to_include": [2, 5]},
            tips=False,
            y_axis_config={"numbers_to_include": [0, 10]},
        )

        labels = ax.get_axis_labels()

        curve_1 = ax.plot(lambda x: 2*x, x_range=[0, 5], color=BLUE_C)
        curve_2 = ax.plot(
            lambda x: 0,
            x_range=[0, 6],
            color=GREEN_B,
        )
        
        t_label = ax.get_graph_label(
            graph=curve_1,
            label= MathTex(r"y=2x"),
            x_val=5,
            dot=False,
            direction=UR)
        
        line_1 = ax.get_vertical_line(ax.input_to_graph_point(2, curve_1), color=YELLOW)
        line_2 = ax.get_vertical_line(ax.i2gp(5, curve_1), color=YELLOW)

#         riemann_area = ax.get_riemann_rectangles(curve_1, x_range=[0.3, 0.6], dx=0.03, color=BLUE, fill_opacity=0.5)
        area = ax.get_area(curve_2, [2, 5], bounded_graph=curve_1, color=YELLOW, opacity=0.5)
    
        transform_title = Tex("Hitung luas daerah yang diarsir?", font_size=72)
        transform_title.to_corner(UP + RIGHT)
        
        hasil_title =MathTex(
            "L = \\frac{1}{2}(a+b) \\times t = \\frac{1}{2}(4+10) \\times 3 = 21 \\text{ satuan luas}"
        )
        
        scene.add(ax, labels, t_label, transform_title)
        
        scene.play(Create(curve_1),run_time=2)
        scene.wait()
        scene.play(Create(curve_2),run_time=2)
        scene.wait()
        scene.play(Create(line_1),run_time=0.5)
        scene.wait()
        scene.play(Create(line_2),run_time=0.5)
        scene.wait()
        scene.play(Create(area),run_time=2.5)
        scene.wait(3)
        scene.play(Create(hasil_title),run_time=5.5)
        scene.wait(3)
        
        scene.play(Uncreate(hasil_title))
        scene.play(Uncreate(area))
        scene.play(Uncreate(line_1))
        scene.play(Uncreate(line_2))
        scene.play(Uncreate(curve_1))
        scene.play(Uncreate(curve_2))
        scene.remove(ax, labels, t_label, transform_title)             
        scene.wait(2)
        
def FifthScene(scene : Scene):
        text = Text("Nah, sekarang bagaimana dengan kurva yang bentuknya tidak beraturan?")
        text.scale(0.5)
        scene.play(Create(text),run_time=2)
        scene.wait(3)
        scene.play(Uncreate(text,reverse=False))
        scene.wait(3)
        
def SixthScene(scene : Scene):
        ax = Axes(
            x_range=[0, 6],
            y_range=[0, 12],
#             x_axis_config={"numbers_to_include": [2,5]},
            tips=False,
#             y_axis_config={"numbers_to_include": [0, 10]},
        )

        labels = ax.get_axis_labels()

        curve_1 = ax.plot(lambda x: (1/3)*x**2, x_range=[0, 5.5], color=RED)
        curve_2 = ax.plot(
            lambda x: 0,
            x_range=[0, 6],
            color=BLUE_B,
        )
        
        x1_label = ax.get_T_label(x_val=2, graph=curve_2, label=Tex("a"))
        x2_label = ax.get_T_label(x_val=5, graph=curve_2, label=Tex("b"))
        
        line_1 = ax.get_vertical_line(ax.input_to_graph_point(2, curve_1), color=YELLOW)
        line_2 = ax.get_vertical_line(ax.i2gp(5, curve_1), color=YELLOW)

#         riemann_area = ax.get_riemann_rectangles(curve_1, x_range=[0.3, 0.6], dx=0.03, color=BLUE, fill_opacity=0.5)
        area = ax.get_area(curve_2, [2, 5], bounded_graph=curve_1, color=BLUE, opacity=0.5)
    
        transform_title = Tex("Misalkan terdapat fungsi y=f(x)", font_size=72)
        transform_title.to_corner(UP + RIGHT)
        
        hasil_title =Tex(
            "Bagaimana menghitung luas di bawah kurva ini?"
        )
        hasil_title.set_color(YELLOW)
        rectangle = BackgroundRectangle(hasil_title, color=RED, fill_opacity=0.5)
        
        scene.add(ax, labels, x1_label, x2_label, transform_title)
        
        scene.play(Create(curve_1),run_time=2)
        scene.wait()
        scene.play(Create(curve_2),run_time=2)
        scene.wait()
        scene.play(Create(line_1),run_time=0.5)
        scene.wait()
        scene.play(Create(line_2),run_time=0.5)
        scene.wait()
        scene.play(Create(area),run_time=2.5)
        scene.wait(3)
        scene.play(Create(hasil_title),run_time=2.5)
        scene.play(Create(rectangle))
        scene.wait(3)
        
        scene.play(Uncreate(rectangle))
        scene.play(Uncreate(hasil_title))
        scene.play(Uncreate(area))
        scene.play(Uncreate(line_1))
        scene.play(Uncreate(line_2))
        scene.play(Uncreate(curve_1))
        scene.play(Uncreate(curve_2))
        scene.remove(ax, labels, x1_label, x2_label, transform_title)             
        scene.wait(2)
        
def SeventhScene(scene : Scene):
        ax = Axes(
            x_range=[0, 6],
            y_range=[0, 12],
#             x_axis_config={"numbers_to_include": [2, 5]},
            tips=False,
#             y_axis_config={"numbers_to_include": [0, 10]},
        )

        labels = ax.get_axis_labels()

        curve_1 = ax.plot(lambda x: (1/3)*x**2, x_range=[0, 5.5], color=RED)
        curve_2 = ax.plot(
            lambda x: 0,
            x_range=[0, 6],
            color=BLUE_B,
        )
        
        x1_label = ax.get_T_label(x_val=2, graph=curve_2, label=Tex("a"))
        x2_label = ax.get_T_label(x_val=5, graph=curve_2, label=Tex("b"))
        
        line_1 = ax.get_vertical_line(ax.input_to_graph_point(2, curve_1), color=YELLOW)
        line_2 = ax.get_vertical_line(ax.i2gp(5, curve_1), color=YELLOW)

        riemann_area6 = ax.get_riemann_rectangles(curve_1, x_range=[2, 5], dx=0.5, color=BLUE, fill_opacity=0.5)
        riemann_area = ax.get_riemann_rectangles(curve_1, x_range=[2, 5], dx=0.05, color=BLUE, fill_opacity=0.5)
#         area = ax.get_area(curve_2, [2, 5], bounded_graph=curve_1, color=GREY, opacity=0.5)
    
        transform_title = Tex("Integral Riemann", font_size=72)
        transform_title.to_corner(UP + RIGHT)
        
        text1 = MathTex(r'\text{Luas tiap 1 persegi} \Rightarrow \;L&=p\;\times\; l \\ &=y\;\times\; \Delta x \\',
                        '&=\sum^{6}_{i=0}f_{L}(i)\Delta x', font_size=34)
        text1.to_corner(UP + LEFT - 0.8)
        
        text2 = MathTex(r'\text{Luas di bawah kurva} \Rightarrow &=\sum^{n\rightarrow \infty}_{i=0}f_{L}(i)\Delta x \\',
                        '&=\int^b_a f(x)dx',font_size=34)
        text2.to_corner(UP + LEFT - 0.8)

        scene.add(ax, labels, x1_label, x2_label, transform_title)
        scene.play(Create(curve_1))
        scene.wait()
        scene.play(Create(curve_2))
        scene.wait()
        scene.play(Create(line_1))
        scene.wait()
        scene.play(Create(line_2))
        scene.wait()
        scene.play(Create(riemann_area6),run_time=2)
        scene.wait()
        scene.play(Write(text1),run_time=5.5)
        scene.wait(3)
        scene.play(Uncreate(riemann_area6))
        scene.play(Unwrite(text1))
        scene.wait(3)
        scene.play(Create(riemann_area),run_time=5.5)
        scene.play(Write(text2),run_time=5.5)
        scene.wait(3)
        
        scene.play(Uncreate(riemann_area))
        scene.play(Unwrite(text2))
        scene.play(Uncreate(line_1))
        scene.play(Uncreate(line_2))
        scene.play(Uncreate(curve_1))
        scene.play(Uncreate(curve_2))
        scene.remove(ax, labels, x1_label, x2_label, transform_title) 
        scene.wait(1)

def EightScene(scene : Scene):
        ax = Axes(
            x_range=[0, 6],
            y_range=[0, 12],
#             x_axis_config={"numbers_to_include": [2, 5]},
            tips=False,
#             y_axis_config={"numbers_to_include": [0, 10]},
        )

        labels = ax.get_axis_labels()

        curve_1 = ax.plot(lambda x: (1/3)*x**2, x_range=[0, 5.5], color=RED)
        curve_2 = ax.plot(
            lambda x: 0,
            x_range=[0, 6],
            color=BLUE_B,
        )

        x1_label = ax.get_T_label(x_val=2, graph=curve_2, label=Tex("a"))
        x2_label = ax.get_T_label(x_val=5, graph=curve_2, label=Tex("b"))
        
        line_1 = ax.get_vertical_line(ax.input_to_graph_point(2, curve_1), color=YELLOW)
        line_2 = ax.get_vertical_line(ax.i2gp(5, curve_1), color=YELLOW)

        riemann_area6 = ax.get_riemann_rectangles(curve_1, x_range=[2, 5], dx=0.5, input_sample_type='right', color=BLUE, fill_opacity=0.5)
        riemann_area = ax.get_riemann_rectangles(curve_1, x_range=[2, 5], dx=0.05,  input_sample_type='right', color=BLUE, fill_opacity=0.5)
#         area = ax.get_area(curve_2, [2, 5], bounded_graph=curve_1, color=GREY, opacity=0.5)
    
        transform_title = Tex("Integral Riemann", font_size=72)
        transform_title.to_corner(UP + RIGHT)
        
        text1 = MathTex(r'\text{Luas tiap 1 persegi} \Rightarrow \;L&=p\;\times\; l \\ &=y\;\times\; \Delta x \\',
                        '&=\sum^{6}_{i=0}f_{R}(i)\Delta x', font_size=34)
        text1.to_corner(UP + LEFT - 0.8)
        
        text2 = MathTex(r'\text{Luas di bawah kurva} \Rightarrow &=\sum^{n\rightarrow \infty}_{i=0}f_{R}(i)\Delta x \\',
                        '&=\int^b_a f(x)dx',font_size=34)
        text2.to_corner(UP + LEFT - 0.8)

        scene.add(ax, labels, x1_label, x2_label, transform_title)
        scene.play(Create(curve_1))
        scene.wait()
        scene.play(Create(curve_2))
        scene.wait()
        scene.play(Create(line_1))
        scene.wait()
        scene.play(Create(line_2))
        scene.wait()
        scene.play(Create(riemann_area6),run_time=2)
        scene.wait()
        scene.play(Write(text1),run_time=5.5)
        scene.wait(3)
        scene.play(Uncreate(riemann_area6))
        scene.play(Unwrite(text1))
        scene.wait(3)
        scene.play(Create(riemann_area),run_time=5.5)
        scene.play(Write(text2),run_time=5.5)
        scene.wait(3)
        
        scene.play(Uncreate(riemann_area))
        scene.play(Unwrite(text2))
        scene.play(Uncreate(line_1))
        scene.play(Uncreate(line_2))
        scene.play(Uncreate(curve_1))
        scene.play(Uncreate(curve_2))
        scene.remove(ax, labels, x1_label, x2_label, transform_title) 
        scene.wait(1)
        
def NinthScene(scene : Scene):
        ax = Axes(
            x_range=[0, 6],
            y_range=[0, 12],
#             x_axis_config={"numbers_to_include": [2, 5]},
            tips=False,
#             y_axis_config={"numbers_to_include": [0, 10]},
        )

        labels = ax.get_axis_labels()

        curve_1 = ax.plot(lambda x: (1/3)*x**2, x_range=[0, 5.5], color=RED)
        curve_2 = ax.plot(
            lambda x: 0,
            x_range=[0, 6],
            color=BLUE_B,
        )

        line_1 = ax.get_vertical_line(ax.input_to_graph_point(2, curve_1), color=YELLOW)
        line_2 = ax.get_vertical_line(ax.i2gp(5, curve_1), color=YELLOW)
        
        x1_label = ax.get_T_label(x_val=2, graph=curve_2, label=Tex("a"))
        x2_label = ax.get_T_label(x_val=5, graph=curve_2, label=Tex("b"))
        
        riemann_area6 = ax.get_riemann_rectangles(curve_1, x_range=[2, 5], dx=0.5, input_sample_type='center', color=BLUE, fill_opacity=0.5)
        riemann_area = ax.get_riemann_rectangles(curve_1, x_range=[2, 5], dx=0.05,  input_sample_type='center', color=BLUE, fill_opacity=0.5)
#         area = ax.get_area(curve_2, [2, 5], bounded_graph=curve_1, color=GREY, opacity=0.5)
    
        transform_title = Tex("Integral Riemann", font_size=72)
        transform_title.to_corner(UP + RIGHT)
        
        text1 = MathTex(r'\text{Luas tiap 1 persegi} \Rightarrow \;L&=p\;\times\; l \\ &=y\;\times\; \Delta x \\',
                        '&=\sum^{6}_{i=0}f_{C}(i)\Delta x', font_size=34)
        text1.to_corner(UP + LEFT - 0.8)
        
        text2 = MathTex(r'\text{Luas di bawah kurva} \Rightarrow &=\sum^{n\rightarrow \infty}_{i=0}f_{C}(i)\Delta x \\',
                        '&=\int^b_a f(x)dx',font_size=34)
        text2.to_corner(UP + LEFT - 0.8)

        scene.add(ax, labels, x1_label, x2_label, transform_title)
        scene.play(Create(curve_1))
        scene.wait()
        scene.play(Create(curve_2))
        scene.wait()
        scene.play(Create(line_1))
        scene.wait()
        scene.play(Create(line_2))
        scene.wait()
        scene.play(Create(riemann_area6),run_time=2)
        scene.wait()
        scene.play(Write(text1),run_time=5.5)
        scene.wait(3)
        scene.play(Uncreate(riemann_area6))
        scene.play(Unwrite(text1))
        scene.wait(3)
        scene.play(Create(riemann_area),run_time=5.5)
        scene.play(Write(text2),run_time=5.5)
        scene.wait(3)
        
        scene.play(Uncreate(riemann_area))
        scene.play(Unwrite(text2))
        scene.play(Uncreate(line_1))
        scene.play(Uncreate(line_2))
        scene.play(Uncreate(curve_1))
        scene.play(Uncreate(curve_2))
        scene.remove(ax, labels, x1_label, x2_label, transform_title) 
        scene.wait(1)
        
def TenthScene(scene : Scene):
        ax = Axes(
            x_range=[0, 6],
            y_range=[0, 12],
            x_axis_config={"numbers_to_include": [1, 4]},
            tips=False,
#             y_axis_config={"numbers_to_include": [0, 10]},
        )

        labels = ax.get_axis_labels()

        curve_1 = ax.plot(lambda x: (1/3)*x**2, x_range=[0, 5.5], color=RED)
        curve_2 = ax.plot(
            lambda x: 0,
            x_range=[0, 6],
            color=BLUE_B,
        )

        line_1 = ax.get_vertical_line(ax.input_to_graph_point(1, curve_1), color=YELLOW)
        line_2 = ax.get_vertical_line(ax.i2gp(4, curve_1), color=YELLOW)

#         riemann_area = ax.get_riemann_rectangles(curve_1, x_range=[0.3, 0.6], dx=0.03, color=BLUE, fill_opacity=0.5)
        area = ax.get_area(curve_2, [1, 4], bounded_graph=curve_1, color=ORANGE, opacity=0.5)
    
        transform_title = Tex("Contoh Soal", font_size=72)
        transform_title.to_corner(UP + RIGHT)
        
        text2 = MathTex(r'L &=\int^b_a f(x)dx \\ &=\int^4_1 \frac{1}{3}x^2 dx \\&=\frac{1}{3 \times 3}(x^3)|^4_1 \\&=\frac{1}{9}((4)^3-1) \\&=7',font_size=34)
        text2.to_corner(UP )
        
        scene.add(ax, labels, transform_title)
        scene.play(Create(curve_1),run_time=2)
        scene.wait()
        scene.play(Create(curve_2),run_time=2)
        scene.wait()
        scene.play(Create(line_1),run_time=0.5)
        scene.wait()
        scene.play(Create(line_2),run_time=0.5)
        scene.wait()
        scene.play(Create(area),run_time=2.5)
        scene.wait(3)
        scene.play(Write(text2),run_time=5.5)
        scene.wait(3)
        
        scene.play(Unwrite(text2))
        scene.play(Uncreate(area))
        scene.play(Uncreate(line_1))
        scene.play(Uncreate(line_2))
        scene.play(Uncreate(curve_1))
        scene.play(Uncreate(curve_2))
        scene.remove(ax, labels, transform_title)             
        scene.wait(2)

def EleventhScene(scene : Scene):
        text1 = Text(
            'Thanks',
            t2c={'[:1]': '#3174f0', '[1:2]': '#e53125',
                 '[2:3]': '#fbb003', '[3:4]': '#3174f0',
                 '[4:5]': '#269a43', '[5:]': '#e53125'}).scale(3)
        scene.play(Write(text1),reverse=True,run_time=2)
        scene.wait(2)
        
class Main(Scene):
         def construct(scene):
            FirstScene(scene)
            SecondScene(scene)
            ThirdScene(scene)
            FourthScene(scene)
            FifthScene(scene)
            SixthScene(scene)
            SeventhScene(scene)
            EightScene(scene)
            NinthScene(scene)
            TenthScene(scene)
            EleventhScene(scene)
            
%manim -qm -v WARNING Main
```
